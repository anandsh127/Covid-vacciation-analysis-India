```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file='vaccination statewise India.csv'
df=pd.read_csv(file)
df['Updated On']= pd.to_datetime(df['Updated On'], format='%d/%m/%Y')

df.info()
df.head()
```

```python
# check weather dataset contain any null values
df.isnull().sum()
 
# replace them with '0'
df.fillna(0,inplace=True)
df.tail()
```

```python
#  dropping unnecessary columns (not required by me for analysis)
df.drop(['Total Sessions Conducted','Total Number of Sites','Male(Individuals Vaccinated)','Female(Individuals Vaccinated)','Transgender(Individuals Vaccinated)'],axis=1,inplace=True)
```

```python
df.columns
```

### Reading data

```python
#the cumulative vaccination uptil the last day given in the data i.e 7th June
Last_updated_day=df[df['Updated On']=='2021-06-07']
Last_updated_day.head()
```

#### population dataset

```python
file='population 2020.csv'
populn_data=pd.read_csv(file)
populn_data.head(10)
```

### Merging both dataframes


#### vaccination data and population data

```python
#---Vaccination & Population data merged---
new_df = pd.concat([Last_updated_day.reset_index(drop=True),populn_data.reset_index(drop=True)], axis=1)

#removing duplicated columns
new_df = new_df.loc[:,~new_df.columns.duplicated()]

#converting datatype from string to float
new_df['Estimated Population 2020(as per UIDAI)'] = new_df['Estimated Population 2020(as per UIDAI)'].str.replace(',', '').astype(float)
new_df.head()
```

### Now some insights

#### 1. Percentage of  people vaccinated wrt it's state population uptil the last day for which data is given
### ie 7th June
   

```python
new_df["%people_vaccinated(atleast_1_dose)"] = new_df["Total Individuals Vaccinated"] / new_df["Estimated Population 2020(as per UIDAI)"]*(100)
new_df=new_df.loc[:,['State','Total Individuals Vaccinated','Estimated Population 2020(as per UIDAI)','%people_vaccinated(atleast_1_dose)']]
new_df=new_df.sort_values(by='%people_vaccinated(atleast_1_dose)',ascending=False).reset_index(drop=True)


def plot_bar(df, x_column, y_column):
    ax = plt.subplots(figsize=(21,10))[1]
    x= df.iloc[:,x_column].to_list()
    y = df.iloc[:,y_column].to_list()
    df_bar= plt.bar(x,y, width=0.6, color= 'yellow')  
    for idx,rect in enumerate(df_bar):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.02*height,
                round(y[idx],2),
                ha='center', va='bottom', rotation=0)
    plt.xticks(rotation=90)
    plt.axhline(y[x.index('India')], color='k',linestyle='--',label="the dotted line going across represents India's average")  ###india pres
    ax.tick_params(axis='x', size=14)
    plt.legend(prop={'size': 14})
    plt.title(df.columns[y_column], size=19)
plot_bar(new_df,0,3)
new_df.head(10)
```

#### --- Ladakh is at no.1 with 49.46% of its people vaccinated(atleast one dose) and UP is at the last with only 7.18% of its people been given first dose. 
#### --- India's average is 13.64%(people vaccinated).

#### 2. To show the first and second doses administered uptil the last day in a stacked bar chart

##### ------ Out of the total doses (first+second) administered, ideally half of them should be the first dose and another half should be the second, in order to achieve the target of giving two doses to each person.

```python
#creating a new dataframe while extracting values from columns of a previous dataframe
Doses=(Last_updated_day.iloc[:,[1,3,4]]).copy()
Doses.head()

Doses['First Dose Administered']=Last_updated_day['First Dose Administered'].div(Last_updated_day['Total Doses Administered'])
Doses['Second Dose Administered'] =Last_updated_day['Second Dose Administered']. div(Last_updated_day['Total Doses Administered'])
                                                                               
Doses.sort_values(by='Second Dose Administered', inplace=True, ascending=False)
Doses.set_index('State', inplace=True)
Doses.plot.bar(stacked= True, figsize=(18,6))
#type(Doses)
Doses.head(10)
```

#### ---- Tripura shows around 30% people of its people got second dose (means both doses) which is 'highest' for a state (wrt its population) in the country.

#### 3. Types of vaccines administered 

```python
grouped=df.groupby('State')
India=grouped.get_group('India')

Covaxin=India['Total Covaxin Administered'].iloc[-2]
Covishield=India['Total CoviShield Administered'].iloc[-2]
SputnikV=India['Total Sputnik V Administered'].iloc[-2]

labels=['Covaxin','CoviShield','SputnikV']
colors=['#42f55a','#f56c42','#008000']

plt.pie([Covaxin,Covishield,SputnikV],colors=colors, labels=labels, autopct='%.2f %%')
plt.title("India's preference")        
plt.show()
```
